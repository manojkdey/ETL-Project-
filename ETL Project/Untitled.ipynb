```python
import pandas as pd
from sqlalchemy import create_engine
csv_file = "vehicles.csv"
vehicles_data_df = pd.read_csv(csv_file)
vehicles_data_df.head()
```

```python
new_vehicles_data_df = vehicles_data_df [['make', 'model', 'trany', 'UCity', 'UHighway', 'year', 'co2TailpipeGpm', 'fuelType', 'drive']].copy()
new_vehicles_data_df.head()
```

```python
cars_country = "countries.csv"
cars_country_df = pd.read_csv(cars_country, encoding="utf-8")
cars_country_df.head()


```

```python
cars_connection_string = "admin:admin@localhost:5432/cars_db"
engine = create_engine(f'postgresql://{cars_connection_string}')
```

```python
engine.table_names()
```

```python
new_vehicles_data_df.to_sql(name='vehicles', con=engine, if_exists='append', index=False)
```

```python
cars_country_df.to_sql(name='cars_by_country', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from vehicles', con=engine).head()
```

```python
pd.read_sql_query('select * from cars_by_country', con=engine).head()
```

```python
#query=
pd.read_sql_query('select  cars_by_country."Country of Origin",vehicles.make, vehicles.model,  vehicles."co2TailpipeGpm"  \
 from vehicles  join cars_by_country  on  cars_by_country."Car Brands" = vehicles.make', con=engine).head()
```
